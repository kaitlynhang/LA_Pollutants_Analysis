```python
# Your code here
#this is just to clean the files up 
def clean_file(df,pollutant):
    if pollutant == 'NO2':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])  
        df = df.rename(columns={'Daily Max 1-hour NO2 Concentration' : 'NO2 Concentration'})
        
    elif pollutant == 'O3':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df = df.rename(columns={'Daily Max 8-hour Ozone Concentration' : 'O3 Concentration'})
    
    elif pollutant == 'PM10':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df = df.rename(columns={'Daily Mean PM10 Concentration' : 'PM10 Concentration'})
        
    elif pollutant == 'PM2.5':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code', 'Method Description' ,'CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df = df.rename(columns={'Daily Mean PM2.5 Concentration' : 'PM2.5 Concentration'})
    
    return df

files= {
    'NO2 LA 2013':'NO2 LA 2013.csv',
    'NO2 LA 2014':'NO2 LA 2014.csv',
    'NO2 LA 2015':'NO2 LA 2015.csv',
    'NO2 LA 2016':'NO2 LA 2016.csv',
    'NO2 LA 2017':'NO2 LA 2017.csv',
    'NO2 LA 2018':'NO2 LA 2018.csv',
    'NO2 LA 2019':'NO2 LA 2019.csv',
    'NO2 LA 2020':'NO2 LA 2020.csv',
    'NO2 LA 2021':'NO2 LA 2021.csv',
    'NO2 LA 2022':'NO2 LA 2022.csv',
    'NO2 LA 2023':'NO2 LA 2023.csv',
    'O3 LA 2013' :'O3 LA 2013.csv',
    'O3 LA 2014' : 'O3 LA 2014.csv',
    'O3 LA 2015' : 'O3 LA 2015.csv',
    'O3 LA 2016' : 'O3 LA 2016.csv',
    'O3 LA 2017' : 'O3 LA 2017.csv',
    'O3 LA 2018' : 'O3 LA 2018.csv',
    'O3 LA 2019' : 'O3 LA 2019.csv',
    'O3 LA 2020' : 'O3 LA 2020.csv',
    'O3 LA 2021' : 'O3 LA 2021.csv',
    'O3 LA 2022' : 'O3 LA 2022.csv',
    'O3 LA 2023' : 'O3 LA 2023.csv',
    'PM10 LA 2013' : 'PM10 LA 2013.csv',
    'PM10 LA 2014' : 'PM10 LA 2014.csv',
    'PM10 LA 2015' : 'PM10 LA 2015.csv',
    'PM10 LA 2016' : 'PM10 LA 2016.csv',
    'PM10 LA 2017' : 'PM10 LA 2017.csv',
    'PM10 LA 2018' : 'PM10 LA 2018.csv',
    'PM10 LA 2019' : 'PM10 LA 2019.csv',
    'PM10 LA 2020' : 'PM10 LA 2020.csv',
    'PM10 LA 2021' : 'PM10 LA 2021.csv',
    'PM10 LA 2022' : 'PM10 LA 2022.csv',
    'PM10 LA 2023' : 'PM10 LA 2023.csv',
    'PM2.5 LA 2013' : 'PM25 LA 2013.csv',
    'PM2.5 LA 2014' : 'PM25 LA 2014.csv',
    'PM2.5 LA 2015' : 'PM25 LA 2015.csv',
    'PM2.5 LA 2016' : 'PM25 LA 2016.csv',
    'PM2.5 LA 2017' : 'PM25 LA 2017.csv',
    'PM2.5 LA 2018' : 'PM25 LA 2018.csv',
    'PM2.5 LA 2019' : 'PM25 LA 2019.csv',
    'PM2.5 LA 2020' : 'PM25 LA 2020.csv',
    'PM2.5 LA 2021' : 'PM25 LA 2021.csv',
    'PM2.5 LA 2022' : 'PM25 LA 2022.csv',
    'PM2.5 LA 2023' : 'PM25 LA 2023.csv',
}
    #made it a dictionary so that I didn't have to keep writing .csv when referencing dataframes
dataframe = {}
for nickname, file_name in files.items(): 
    pollutant = nickname.split()[0]
    df = pd.read_csv(file_name, index_col='Date') #Will read the CSV file into a Dataframe and set index to Dates
    if df.index.duplicated().any(): #Will check for duplicates in the index column
        df = df[~df.index.duplicated(keep='first')] #Will remove the duplicates
    df_cleaned = clean_file(df,pollutant)
    dataframe[nickname] = df_cleaned
  


print(dataframe['PM2.5 LA 2015'])
```

```python

#Initializing variables
mean_values = {}
std_values = {}
median_values = {}
first_quartile_values = {}
third_quartile_values = {}
units = {}

for nickname, df in dataframe.items():
        pollutant_type = nickname.split()[0] #getting the type of pollutant from the nickname
        concentration_column = f'{pollutant_type} Concentration' #getting the concentration in the concentration column
        if concentration_column in df.columns: #Getting statistics values
            mean_values[nickname] = df[concentration_column].mean()
            std_values[nickname] = df[concentration_column].std()
            median_values[nickname] = df[concentration_column].median()
            first_quartile_values[nickname] = df[concentration_column].quantile(0.25)
            third_quartile_values[nickname] = df[concentration_column].quantile(0.75)
            units[nickname] = df['Units'].iloc[0] #if 'Units' in df.columns else 'N/A'

summary_df = pd.DataFrame({
    'Molec and Year': list(mean_values.keys()),
    'Mean': list(mean_values.values()),
    'Units': list(units.values()),
    'Std Deviation': list(std_values.values()),
    '1st Quartile': list(first_quartile_values.values()),
    'Median': list(median_values.values()),
    '3rd Quartile': list(third_quartile_values.values())
}) 
#I put it into another dataframe so I could make a summary table

summary_df.set_index('Molec and Year', inplace=True) #just setting the index so the table looks nicer
print(summary_df)

print('-' * 115)

pollutant_data = []
#Just for labeling the axes
Units = {
    'NO2': 'ppb',  
    'O3': 'ppb',   
    'PM10': 'g/m SC', 
    'PM2.5': 'g/m LC' 
}

for nickname, df in dataframe.items():
    pollutant_type = nickname.split()[0] #getting the type of pollutant from nickname
    year = nickname.split()[2] #getting the year from nickname
    concentration_column = f'{pollutant_type} Concentration'
    
    if concentration_column in df.columns:
        df_cleaned = df.copy()  # Avoid modifying the original DataFrame
        df_cleaned['Year'] = year
        df_cleaned['Pollutant'] = pollutant_type
        df_cleaned = df_cleaned[['Year', 'Pollutant', concentration_column]]
        df_cleaned = df_cleaned.rename(columns={concentration_column: 'Concentration'})
        pollutant_data.append(df_cleaned)

# Combinining all data into a dataframe
combined_df = pd.concat(pollutant_data)

pollutants = combined_df['Pollutant'].unique()

#Box and Whisker Plot Code
fig, axes = plt.subplots(4, 1, figsize=(15, 30), sharey=False)

# Flattening the axes array 
axes = axes.flatten()

for i, pollutant in enumerate(pollutants):
    ax = axes[i]
    data_to_plot = [combined_df[(combined_df['Pollutant'] == pollutant) & (combined_df['Year'] == year)]['Concentration'].dropna() for year in sorted(combined_df['Year'].unique())]
    ax.boxplot(data_to_plot, labels=sorted(combined_df['Year'].unique()))
    ax.set_title(f'{pollutant} Concentration Across Years')
    ax.set_xlabel('Year')
    ax.set_ylabel(f'Concentration ({Units[pollutant]})')
    ax.tick_params(axis='x', rotation=45)
    
plt.tight_layout()
plt.show()
```

### 1.4: Time series analysis

Let's explore how pollutant concentrations change over time for each of the 10 years in the dataset.

Your goal is:
1. To make a visual(s) that clearly shows the variation in pollutant concentrations as a function of day of the year. Your visual should communicate any differences during the lockdown period compared to "normal" years

<code style="background:green;color:white">2. Do each of the years follow the same general seasonal trend? Are there particular years that stand out- in what way? Are the seasonal trends you observe for the larger dataset similar to what you observed for the 2019-2021 dataset or different? How?> 

```python
# Your code here
#Initializing variables
mean2_values = {}
std2_values = {}
median2_values = {}
first2_quartile_values = {}
third2_quartile_values = {}
processed_month = set() #to make sure that there are no duplicates

#created a for loop so that I can iterate through each file and do its stats
for nickname, df in dataframe.items():
    pollutant_type = nickname.split()[0]
    concentration_column = f'{pollutant_type} Concentration'
    if concentration_column in df.columns:
        df.index = pd.to_datetime(df.index)
        df = df.sort_index()
        grouped_by_month = df.groupby(df.index.month) #grouping data by month
        for month, group in grouped_by_month:
            mean2_values[(nickname, month)] = group[concentration_column].mean()
            std2_values[(nickname, month)] = group[concentration_column].std()
            median2_values[(nickname, month)] = group[concentration_column].median()
            first2_quartile_values[(nickname, month)] = group[concentration_column].quantile(0.25)
            third2_quartile_values[(nickname, month)] = group[concentration_column].quantile(0.75)
            processed_month.add(month)

#made another DataFrame so I can print it as a table
summary2_df = pd.DataFrame({
    'Molec and Year': list(mean2_values.keys()),
    'Mean': list(mean2_values.values()),
    'Std Deviation': list(std2_values.values()),
    '1st Quartile': list(first2_quartile_values.values()),
    'Median': list(median2_values.values()),
    '3rd Quartile': list(third2_quartile_values.values())
}) 

summary2_df.set_index('Molec and Year', inplace=True)
#pd.set_option('display.max_rows', None)  #Can uncomment if you want to see all the pollutant, year, and month
print(summary2_df)

print('-' *105)

Units = {
    'NO2': 'ppb',  
    'O3': 'ppb',   
    'PM10': 'g/m SC', 
    'PM2.5': 'g/m LC' 
}

#defining variables and having a place to store
x_values = {}
y_values = {}

for pollutant in ['NO2', 'O3', 'PM10', 'PM2.5']: #A for loop to iterate over each different pollutant
    x_values[pollutant] = {}
    y_values[pollutant] = {}
    for year in ['2013','2014','2015','2016','2017','2018', '2019', '2020', '2021','2022','2023']: #A for loop to iterate over each year for each different pollutant
        x_values[pollutant][year] = []
        y_values[pollutant][year] = []
        for month, mean_value in mean2_values.items(): #A for loop to iterate over each month and get its mean 
                                                        #value after it iterated each pollutant and year
            if month[0].startswith(pollutant) and month[0].endswith(year): #was having problems with matching, 
                                                                            #so this was a check
                x_values[pollutant][year].append(month[1])  
                y_values[pollutant][year].append(mean_value)  

#Below is the code for the plots

for pollutant in ['NO2', 'O3', 'PM10', 'PM2.5']: #a for loop to create a new graph for each pollutant
    plt.figure()
    # Plot each year's data on the same subplot
    for year, color in zip(['2013','2014','2015','2016','2017','2018', '2019', '2020', '2021','2022','2023'], ['red', 'green', 'blue','purple','orange','red','yellow','black','pink','lightblue','lightgreen']): #a for loop to make sure that
                                                                                #each year got a diff color per
                                                                                #pollutant graph
        plt.plot(x_values[pollutant][year], y_values[pollutant][year], label=f'{pollutant} {year}', color=color, marker='o')
    plt.xlabel('Month')
    plt.ylabel(f'Concentration ({Units[pollutant]})')
    plt.title(f'Mean Concentration of {pollutant} Across Different Months and Years')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)

plt.show()
```

```python
# your code here

def clean_file_SD(df,pollutant):
    if pollutant == 'NO2':
        df = df.drop(columns=['Site ID','POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])  
        df = df.rename(columns={'Daily Max 1-hour NO2 Concentration' : 'NO2 Concentration'})
        
    elif pollutant == 'O3':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df = df.rename(columns={'Daily Max 8-hour Ozone Concentration' : 'O3 Concentration'})
    
    elif pollutant == 'PM10':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df= df.rename(columns={'Daily Mean PM10 Concentration' : 'PM10 Concentration'})
        
    elif pollutant == 'PM2.5':
        df = df.drop(columns=['Site ID', 'POC','Source','Daily Obs Count','Percent Complete','AQS Parameter Code','AQS Parameter Description','Method Code','Method Description','CBSA Code','CBSA Name','State FIPS Code','County FIPS Code','County','Site Latitude','Site Longitude'])
        df= df.rename(columns={'Daily Mean PM2.5 Concentration' : 'PM2.5 Concentration'})
    
    return df
    
files_SD= {
    'NO2 SD 2019':'NO2 SD 2019.csv',
    'NO2 SD 2020':'NO2 SD 2020.csv',
    'NO2 SD 2021':'NO2 SD 2021.csv',
    'O3 SD 2019' : 'O3 SD 2019.csv',
    'O3 SD 2020' : 'O3 SD 2020.csv',
    'O3 SD 2021' : 'O3 SD 2021.csv',
    'PM10 SD 2019' : 'PM10 SD 2019.csv',
    'PM10 SD 2020' : 'PM10 SD 2020.csv',
    'PM10 SD 2021' : 'PM10 SD 2021.csv',
    'PM2.5 SD 2019' : 'PM25 SD 2019.csv',
    'PM2.5 SD 2020' : 'PM25 SD 2020.csv',
    'PM2.5 SD 2021' : 'PM25 SD 2021.csv',
}
    #made it a dictionary so that I didn't have to keep writing .csv when referencing dataframes
dataframe_SD = {}
for nickname, file_name in files_SD.items(): 
    pollutant = nickname.split()[0]
    df = pd.read_csv(file_name, index_col='Date') #Will read the CSV file into a Dataframe and set index to Dates
    if df.index.duplicated().any(): #Will check for duplicates in the index column
        df = df[~df.index.duplicated(keep='first')] #Will remove the duplicates
    df_cleaned_SD = clean_file_SD(df,pollutant)
    dataframe_SD[nickname] = df_cleaned_SD
  


print(dataframe_SD['NO2 SD 2020'])
```
